```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import fitz
from bs4 import BeautifulSoup

# Carregamento CSV
df_csv = pd.read_csv("modelo1a.csv")
text_csv = df_csv['NOTA FISCAL – Modelo 1-A'].tolist()
labels_csv = df_csv['NOTA FISCAL – Modelo 1-A'].tolist()

from google.colab import files
uploaded_pdf = files.upload()
# Extrair texto
def extrair_texto_pdf(arquivo_pdf):
    textos = []
    doc = fitz.open(arquivo_pdf)

    for pagina in doc.pages():
        texto = pagina.get_text()
        textos.append(texto)

    return textos

# Carregamento PDF
text_pdf = extrair_texto_pdf("modelo1a.pdf")

from google.colab import files
from bs4 import BeautifulSoup

uploaded_html = files.upload()

# Extrair texto  HTML
def extrair_texto_html(arquivo_html):
    textos = []

    # Verificar se o HTML foi carregado
    if arquivo_html:
        soup = BeautifulSoup(arquivo_html, 'html.parser')
        texto = soup.get_text()
        textos.append(texto)

    return textos

# Nome do arquivo HTML
html_file = next(iter(uploaded_html), None)

text_html = extrair_texto_html(uploaded_html[html_file] if html_file else None)

# Combinação de textos
textos = text_csv + text_pdf + text_html
labels = labels_csv + ["PDF"] * len(text_pdf) + ["HTML"] * len(text_html)

textos = ["" if pd.isna(t) else t for t in textos]

# Pré-processamento dos dados
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(textos)


# Divisão dos dados
X_train, X_test, y_train, y_test = train_test_split(X_train_vectorized, labels, test_size=0.2, random_state=42)

# Treina o modelo
model = SVC()
model.fit(X_train, y_train)

# Performance
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Reconhecimento do modelo: {accuracy:.2f}")
print(f"Relatório de Classificação:")
print(classification_report(y_test, predictions, zero_division=1))
```
